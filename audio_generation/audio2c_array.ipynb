```python
import numpy as np

fp = "/content/drive/MyDrive/ECSE444/gameover.mp3"  # change to the correct path to your file accordingly
from audio2numpy import open_audio

signal, sampling_rate = open_audio(fp)

# Average both channels to create a mono signal
mono_signal = np.mean(signal, axis=1)

# Normalize the signal and convert it to uint16_t while storing 12-bit values
normalized_signal = ((mono_signal + 1) * ((2**12 - 1) / 2)).astype(np.uint16)

# Save the signal and sampling rate to a header file
header_file = "/content/drive/MyDrive/ECSE444/gameover_signal.h"
with open(header_file, "w") as f:
    f.write("#ifndef AUDIO_SIGNAL_H\n")
    f.write("#define AUDIO_SIGNAL_H\n\n")
    f.write(f"#define SAMPLING_RATE {sampling_rate}\n")
    f.write(f"#define SIGNAL_LENGTH {len(normalized_signal)}\n\n")
    f.write("const uint16_t audio_signal[SIGNAL_LENGTH] = {")
    f.write(", ".join(map(str, normalized_signal)))
    f.write("};\n\n")
    f.write("#endif // AUDIO_SIGNAL_H\n")
```
